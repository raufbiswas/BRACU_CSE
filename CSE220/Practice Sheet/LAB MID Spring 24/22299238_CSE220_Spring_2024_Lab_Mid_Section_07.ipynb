Sample Input:
Stack: 10, 3, 50, 5, 8, 1, 45
(*Topmost element is 45*)

Sample Output:

(*Topmost element is 50*)

Explanation:
50 is the largest number which is at the top of the stack, while 1 is the smallest number and it is at the bottom of the stack.

# Question 2: **Add two Linked List**


You are given two linked lists. Insert the second linked list in between a certain index of the first linked list.

Sample Input:\
head1: A -> B -> F -> G\
head2: C -> D -> E\
index: 2

Sample Output:
head: A -> B -> C -> D -> E -> F -> G





```python
#Run this cell before running your code
import numpy as np
class Node:
  def __init__(self,elem,next = None):
    self.elem,self.next = elem,next

def createList(arr):
  head = Node(arr[0])
  tail = head
  for i in range(1,len(arr)):
    newNode = Node(arr[i])
    tail.next = newNode
    tail = newNode
  return head

def printLinkedList(head):
  temp = head
  while temp != None:
    if temp.next != None:
      print(temp.elem, end = '-->')
    else:
      print(temp.elem)
    temp = temp.next
  print()
```

```python
#helper function  1
def nodeAt(head, idx):
  temp = head
  count = 0
  obj = None
  while temp != None:
    if count == idx :
      obj = temp
      break
    temp = temp.next
    count += 1
  if obj == None:
    print("invalid Index!")

  return obj

#helper Function 2
def nodeCounter(head):
  count = 0
  temp = head
  while temp != None:
    count += 1
    temp = temp.next
  return count
```

```python
def merge_LL(head1, head2, index):
    add_idx = index - 1
    needed_node = nodeAt(head1, add_idx)
    nextOf_needed = needed_node.next
    size1 = nodeCounter(head1)
    size2 = nodeCounter(head2)
    if add_idx < size1:
      needed_node.next = head2
      new_count = nodeCounter(head1)
      last_1 = nodeAt(head1, new_count-1)
      last_1.next = nextOf_needed
      return head1



print('==============Test Case 1=============')
head_1 = createList(np.array(['A', 'B', 'F', 'G']))
head_2 = createList(np.array(['C', 'D', 'E']))
index = 2
print('Link List 1: ', end = ' ')
printLinkedList(head_1)
print('Linked List 2: ', end = ' ')
printLinkedList(head_2)
head = merge_LL(head_1, head_2, index)
print('Merged List: ', end = ' ')
printLinkedList(head) # This should print  A -> B -> C -> D -> E -> F -> G

print('==============Test Case 2=============')
head_1 = createList(np.array(['N', 'O', 'R', 'T', 'H', 'R', 'E', 'S']))
head_2 = createList(np.array(['M', 'E', 'M', 'B', 'E', 'R']))
index = 7
print('Link List 1: ', end = ' ')
printLinkedList(head_1)
print('Linked List 2: ', end = ' ')
printLinkedList(head_2)
head = merge_LL(head_1, head_2, index)
print('Merged List: ', end = ' ')
printLinkedList(head) # This should print  N->O->R->T->H->R->E->M->E->M->B->E->R->S
```
