**Built-in functions for Sorting**

```python
####============= sorted() ===========
number = [-1,-2,-3,4,5,7,-4,3,2,5,6]
numbers_sorted = sorted(number)
numbers_sorted_reverse = sorted(number, reverse = True)
print(number)
print(numbers_sorted)
print(numbers_sorted_reverse)
```

**Selection Sort**

```python
#Ascending Order
def selection_sort(numbers):
 for index in range(len(numbers)-1):
    min = numbers[index]
    min_index = index
    #find minimum and swap with numbers[index]
    for j in range(index+1, len(numbers)):
         if numbers[j] < min:
               min = numbers[j]
               min_index = j 
    numbers[min_index], numbers[index] = numbers[index], numbers[min_index]   
 return numbers

numbers = [3,5,7,1,2,-11,-66,4,100, 99]
sorted_numbers = selection_sort(numbers)
print(sorted_numbers)
```

**Bubble Sort**

```python
def bubble_sort(numbers):
  for i in range(0, len(numbers)-1):
    for j in range(0,len(numbers)-i-1):
      if numbers[j]> numbers[j+1]:
        numbers[j], numbers[j+1] = numbers[j+1], numbers[j]
  return numbers

bubble_sorted = bubble_sort([5,3,1,9,8,2,4,7])
print("Sorted:", bubble_sorted)
```

**Linear Search**

```python
def linear_search(data, x):
    for index in range(len(data)):
         if x == data[index]:
             return index #index if found
    return False

position = linear_search([2,4,5,6,7,8,3,7,1], 4)

if position == False:
  print('Value is not found !!')
else:
  print("Value found in index no:", position)
```

**Binary Search**

```python
def binary_search(data, x):

    left = 0
    right = len(data) - 1
    while left <= right:
        mid_index = (left+ right)//2
        mid_val = data[mid_index]
        if x == mid_val:
            return mid_index
        elif x > mid_val:
            left = mid_index + 1
        else:
            right = mid_index - 1

    return -1 #sentinel if not found

data = [-5, 5, 10, 15, 20, 50, 100, 150, 200, 300]
index = binary_search(data, 500)

if index == -1:
  print("Value not found !!")
else:
  print("Value found in index no:", index)
```
