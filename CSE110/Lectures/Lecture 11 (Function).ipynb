**function parts**

* **Function name:** It can be anything as we follow naming conventions. Function has the same naming convension as variables

* **Parameter List:** Optional. We use it if we want to pass some values to the function

* **Body of the Function:** Optional. The block of code that is executed when the function is called

* **Return statement:** Optional. After the function has finished executing it will take away a value and replace the function calling statement with that values

* **Calling the function:** A function only executes if we call it. We can call a function following this syntax:

    Function_name(arguments)

```python
#(x,y) - Parameter list 
# x y - Arguments

def compare(x,y):
  max2 = None
  if x>y:
    max2 = x
    print(max2)
  else:
    max2 = y
  print(max2)
  return max2
  print(max2)
  print('\\\\')

s = compare(20,30)
print(s)
```

```python

def hello(x,y):
  max = None
  if x>y:
    max = x
  else:
    max = y
  return max

hello(20,30)
```

**Parameter and Arguments**

* Information can be passed into functions are arguments
* Arguments are specified as parameter list after the function name, inside the parentheses. You can add as many parameters as you want, just searate then with a comma (,)

From a function's perspective:
   * A parameter is the variable listed inside the parentheses in the function definition.
   * A argument is the value that is sent to the function when it is called.

   

```python
def my_function(f_name):
  print("Hello", f_name)

my_function('Alice')
my_function('Max')
my_function('Bob')
```

**Return Statement**

* To let a fuction **return** a value, use the **return** statement
* The returned value replaces the function calling statement

```python
def multiply(x):
  return 5*x

print('Value =', multiply(2))
print('Value =', multiply(2)+multiply(5))
```

**Unknown Number of Arguments**

* If we don't know how many arguments will be passed to the fuction, we can add an asteric(*) before the parameter name in the function definition.

```python
def summation(*number):
  print(number)
  sum1 = 0
  for a in number: 
    sum1+=a  

total_sum = summation(1,2,3)
print(total_sum) 
#print('Total sum is',total_sum)
# total_sum = summation(1,2,3,5,6,7,9)
#print('Total sum is',total_sum)
```

* This way the order of the arguments does not matter

**Unknown number of keyword Arguments** (**kwargs)

* If the number of keyword arguments is unknown, add a double asteric(**) before the parameter name

```python
def my_function(**car):
  print(car)
  print('Car name is', car['name'], 'and model is', car['model'], car['age'])


#my_function(name = 'Ferrari', model = 'A')
my_function(name = 'Lamborghini', model = 'B', color = 'red', age = 4, a =34)
```

**Default Arguments Values**

* In functions parameter list we can specify a default values for one or more arguments
* A default value can be written in the format "parameter_name = value"
* We will have the option to assign or not a value for those arguments

```python
def calculate_sqauresum(x = 2,y = 3):
  print(x)
  print(y)

  return (x*x + y*y )
  print('Hello')

print("The square of the sum of 2 and 5 is :", calculate_sqauresum())
#print("The square of the sum of 2 and 2 is :", calculate_sqauresum(3))
#print("The square of the sum of 2 and 5 is :", calculate_sqauresum(2, 5))
```
