**Read the value N from the user, and print the first Nth Fibonacci numbers.**



```python
term = int(input("Enter a value:"))
first = 0
second = 1
fib = 0
# if term<0:
#   print("Invalid input")
# elif term==0:
#   print("No series")
# else:
#       print("Fibonacci Series:", end = " ")
while fib<term:
  if fib == 0:
    print(fib, end=' ')
  first = second
  second = fib
  fib = first + second
  if fib<term:
    print(fib,end = ' ')
```

Write a Python code that will calculate the **value of y if the expression** of y is as follows (n is the input):  

$y = 1^2 - 2^2 +3^2 -4^2 +5^2 .........+ n^2 $

```python
num = int(input("Enter a number:"))
sum = 0
for a in range(1,num+1):
  if a%2 != 0:
    sum+=a**2
  elif a%2==0:
    sum-=a**2
print(sum)
```
