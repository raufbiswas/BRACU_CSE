**Tuple**


* A tuple is a collection of objects which is ordered and unchangeable/immutable.  
* Tuples are used to store multiple items in a single variable  
* Tuples are sequences just like lists. But the differences are between tuple and list are:
  * Lists are mutable/changable whereas tuples are immutable
  * Lists use square brackets [ ] but tuples use parentheses ( )
* Example:
> my_tuple = ("apple", "banana", "cherry")

```python
val = (1,)
print(tuple(val))
print(type(val))
```

**Accessing Elements of a Tuple**
* It has both positive and negative indexing
* Index must be an integer

```python
#Accesing tuple elements using indexing

my_tuple = (1,2,'World',True, 5, 'r','s')
print(my_tuple[0])
print(my_tuple[3])
print(my_tuple[-5][2])
print(my_tuple[-2])
```

```python
#Accesing tuple elements from nested tuple

my_tuple = ("house", 5, [8, 4, 6], (1, 2, 3))
print(my_tuple[0][-3])
print(my_tuple[2][1])
print(my_tuple[3][1])
print(my_tuple[-2][2])
```

**Mutability of Tuples**

* Tuples are immutable. Elements of tuples can't be changed once they have been assigned
* If the element is itself a mutable data like list then it's nested items can be changed

```python
# Mutability check
tuple1 = (1,2,4,'Hello')
my = list(tuple1)
print(my)
my[3]='World'
print(my)
tuple1 = tuple(my)
print(tuple1)
```

**index()**


*   Searches for the first occurrence of the value, and returns its position.



```python
tuple_1 = (53,32,78,12,2,3,6,7,53,53, [1,2])
x = int(input())
for a in range(len(tuple_1)):
  if tuple_1[a] == x:
    print('Index =', a)
```

**enumerate()**


*   It takes a list as a parameter and returns a tuple for each element in the list.




```python
grocery = ['bread', 'milk', 'butter', [1,2,4]]
newGrocery = enumerate(grocery)
# print(newGrocery)
# converting to tuple
print(tuple(newGrocery))



  
```
