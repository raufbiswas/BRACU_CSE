**Dictionary**



*  Dictionary is a data structure to hold unordered collection of values in a single variable  
*  Each item in a dictionary has a **key:value** pair
*  It can have any number of items with a unique key 
*  A dictionary is mutable (can change individual items)






**Syntax:**

my_dict = {key1:value1, key2:value2, key3:value3}



*   All the items are enclosed within the curly braces { }
*   Commas (,) to separate each item
* Colon(:) to separate keys from its values
* Keys are unique and can be any immutable data type (strings/numbers/tuples)
* Values do not have to be unique and can be any data type including dictionaries.




**Creating Dictionaries**

A dictionary can be created either using the curly braces {} or by using built-in function, dict()


```python
my_dict = {'Name': 'Max', 'Age': 24, 'Country': {'Spain':2, 3:'Hello'}}
print(type({'Spain':2, 3:'Hello'}))
for a in my_dict.values():
    for b,c in a.items():
      print(b)
      print(c)
```

```python
my = {1:2}
new = {3:4}
for a,b in new.items():
  my[a]=b
print(my)
```

**Accessing Dictionary Values**



 1. Just like strings and lists, can use square brackets but keys must be used
instead of index
* If **no valid key** is used then it will give **KeyError**

2.  Can use built-in function **get(key,[value])**

  * Key: to be searched
  * Value: Optional. If no value found, the set value will be returned. If not set, will return None



```python
# Using square bracket

my_dict = {'Fruit':'Apple', 'Flower': 'Lily', 'Animal': ['Lion', 'cat']}
#print(my_dict['Fruit'])
#print(my_dict['Animal'])
print(my_dict['Animal'][1])

# This will give an error

my_dict = {1:2,3:'Hello'}
print(my_dict)
```

Using traditiona square brackets to add or update the key : value pairs

```python
#Updating The value of Age
my_dict = {'Name': 'Max', 'Age': 24, 'Country': 'Spain'}
print(my_dict['Age'])
my_dict['Age']= 15
print(my_dict['Age'])
print(my_dict)
```

**Deleting Elements**

clear() - removes all the items from the dictionary

popitem() - removes and return the last added kye:value pair. If dictionary is empty return KeyError

pop(key[,msg]) - msg is optional. Returns value and removes key:value pair. If key not found and msg provided will return the msg or will return KeyError


```python
my_dict = {'Name': 'Max', 'Age': 24, 'Country': 'Spain'}
my_dict.clear()
print(my_dict)


```

```python
my_dict = {'Name': 'Max', 'Age': 24, 'Country': 'Spain'}
print(my_dict['Country'])
print(my_dict.popitem())
print(my_dict)
```

**dictionary.keys()**

Does not take any parameters. Returns a new object that contains a list of all keys in the dictionary

```python
my_dict = {'Name': 'Max', 'Age': 24, 'Country': 'Spain'}
for i,j in my_dict.items():
  print(i)
  print(j)
```

**Dictionary Iteration**

**Iterating over keys**

```python
my_dict = {'Name': 'Max', 'Age': 24, 'Country': 'Spain'}

for nazia in my_dict:
  print(nazia)
```

```python
my_dict = {'Name': 'Max', 'Age': 24, 'Country': 'Spain'}

for nazia in my_dict.keys():
  print(nazia)
```

**Iterating over values**

```python
my_dict = {'Name': 'Max', 'Age': 24, 'Country': 'Spain'}

for value in my_dict.values():
  print(value)
```

**Iterating over key and values**

```python
my_dict = {'Name': 'Max', 'Age': 24, 'Country': 'Spain'}

for x,y in my_dict.items():
  print(x, y)
```

**Copying a dictionary**

```python
# copying a dictionary
info = {'Tony': 50, 'Steve': 105, 'Natasha': 30, 'Bruce': 45, 'Thor': 1500}
new_info = {}
print(new_info)
new_info = info.copy()
print(new_info)

# You can check if the two dictionaries are same or not.
# id() function returns an unique identifier (reference/location)
# print(id(info))
# print(id(new_info))
# id() returns different values, so copied successfully

my = [1,23,3]
my1 = my.copy()
# print(id(my))
# print(id(my1))
my1 = my
print(id(my))
print(id(my1))
```
