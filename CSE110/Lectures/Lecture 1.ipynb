**Variables**


* Variables must have a meaningful name
* Variable name must start with small letter
* Variable names can have letters (A-Z),(a-z), digits (0-9) and underscores ‘_’
* Use ‘_’ between words in a variable 
* Case sensitive




```python
name = 'Bob'
```

```python
student_1 = 'Max'
student2 = 'Bob'
```

```python
student_ID = 18101006
```

```python
full_name = 'Nazia Iqbal'
```

* Variable names can not have whitespaces and special characters (+,-,!,@,$,#,%)

```python
$name = 'Bob'      #This is an inappropriate variable name
```
