**SET A**

```python
value = input()
my_dict = {}
categories = ['Letters', 'Digits', 'Special Characters']
for a in value:
  if a!= ' ':
    if a.isupper() == True or a.islower() == True:
      if 'Letters' not in my_dict.keys():
           my_dict['Letters'] = [a]
      else:
          my_dict['Letters'].append(a)

    elif a.isdigit() == True:
      if 'Digits' not in my_dict.keys():
           my_dict['Digits'] = [int(a)]
      else:
          my_dict['Digits'].append(int(a))
    else:
      if 'Special Characters' not in my_dict.keys():
           my_dict['Special Characters'] = [a]
      else:
          my_dict['Special Characters'].append(a) 

for b in categories:
  if b not in my_dict:
    my_dict[b] = []
  else:
    my_dict[b] = tuple(my_dict[b])

print(my_dict)     
```

```python
def less_than_average(my_list):
  length = 0
  new_list = []
  for a in my_list:
    length+=len(a)
  avg_len = length/len(my_list)
  for b in my_list:
    if len(b)< avg_len:
      new_list.append(b)
  return avg_len, new_list

print(less_than_average(["CSE111's","Final","Examination","is","extremely","easy"]))
```
