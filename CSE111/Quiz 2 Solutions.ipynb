```python
class BracuScholarship:
  def __init__(self, name, cgpa, credit):
    self.name = name
    self.cgpa = cgpa
    self.credit = credit
    print(f'{self.name} has applied for scholarship')
  
  def printInfo(self):
    print(f'Name: {self.name}\nCGPA: {self.cgpa}\nCompleted Credit: {self.credit}')
  
  def scholarshipCheck(self, enrolledCourse):
    self.enrolledCourse = enrolledCourse
    #print(f'{self.name} took {self.enrolledCourse} course/s this semester!')

    if self.enrolledCourse*3 >=12 and self.cgpa>=3.80 and self.credit>=30:
      return f'{self.name} took {self.enrolledCourse} course/s this semester!\nCongratulations! You are eligible for scholarship!!'
    else:
      return f'{self.name} took {self.enrolledCourse} course/s this semester!\nSorry! You are not eligible for scholarship'

  
std1 = BracuScholarship('Mahmud', 3.45, 36)
print('------------------------------------')
std1.printInfo()
print('####################################')
print(std1.scholarshipCheck(3))
std2 = BracuScholarship('Zerin', 3.82, 40)
print('------------------------------------')
std2.printInfo()
print('####################################')
print(std2.scholarshipCheck(4))
std3 = BracuScholarship('Raisa', 3.82, 15)
print('------------------------------------')
```

```python
class Library:

  def __init__(self, name, val):
    self.name = name
    # self.d = val[0:-3]
    # self.i = val[-2:]
    self.new = val.split('-')
    self.dept = self.new[0]
    self.id = self.new[1]

    print('A new member is added')

  def memberInfo(self):
    return f'Name: {self.name}\n' f'Department: {self.dept}\n' f'ID: {self.id}'
  
  def addBooks(self, book1, book2):
    self.book1 = book1
    self.book2 = book2
    self.bookNames = {self.name: [self.book1, self.book2]}
    return self.bookNames


member1 = Library('Zulkarnain', 'CSE-01')
print('----------------------------------')
print(member1.memberInfo())
print(member1.addBooks('The Magic Tree', 'Wizards of Ice'))
member2 = Library('Tahphim', 'MNS-04')
print('----------------------------------')
print(member2.memberInfo())
print(member2.addBooks('The Dark Dragon', 'The Last Flame'))
member3 = Library('Faiza', 'CSE-02')
print('----------------------------------')
print(member3.memberInfo())
```
