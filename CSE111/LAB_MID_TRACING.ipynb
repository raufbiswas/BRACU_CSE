```python
class MidTerm:
    def __init__(self, *args):
        self.x = args[1]
        self.y = args[0]-args[2]
        self.z = args[0] * 3
        print(self.y, self.z)
    def methodA(self, x, y = None):
        self.x -= x
        self.z = self.x + self.y
        z = self.methodB({14:6})
        for key in z:
            print(key, z[key])
        if y == None:
            print(self.y, self.z)
    def methodB(self, dict1):
        dict1[self.x] = self.z
        a = self.x - self.z
        dict1[self.x] = a
        dict1[self.y-self.z] = a*2
        return dict1
exam = MidTerm(2,4,9)
exam.methodA(7)
```
